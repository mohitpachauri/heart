### IMPORTING PACKAGES ###

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from matplotlib.pyplot import style ; style.use('ggplot')
```

### TRAVERSING DATA ###

```python
df = pd.read_csv('cancer (1).csv')
df.head(10)
```

#### VALUE COUNT ####

```python
df.diagnosis.value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':0,'B':1})
```

```python
x = df.iloc[:,2:]
y = df.iloc[:,1]
```

```python
x.head()
```

#### COUNT PLOT FOR CHECKING IM-BALANCED DATASET ####

```python
sns.countplot(y)
```

#### CORRELATION BETWEEN ATTRIBUTES #####

```python
x.corr()
```

```python
df2 = df.iloc[:,1:]
```

```python
df2.corr()
```

```python
plt.figure(figsize=(18,16))
sns.heatmap(df2.corr(),annot=True,cmap='RdYlGn')
plt.show()
```

#### DATA REDUCTION ####

```python
from sklearn.decomposition import PCA
pca = PCA(2)

pca.fit(x,y)
x_pca = pca.transform(x)
```

```python
plt.figure(figsize=(16,14))
sns.scatterplot(x_pca[:,0],x_pca[:,1],hue=y)
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(2)
x_tsne = tsne.fit_transform(x,y)
```

```python
plt.figure(figsize=(16,14))
sns.scatterplot(x_tsne[:,0],x_tsne[:,1],hue=y)
plt.show()
```

#### APPLYING ALGORITHMS ####

```python
from sklearn.svm import SVC
svm = SVC(gamma=0.01)
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train , y_test = tts(x,y)
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier as RF
rf = RF()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier as GF
gf = GF()
gf.fit(x_train,y_train)
gf.score(x_test,y_test)
```

```python
from xgboost import XGBClassifier as XG
xg = XG()
xg.fit(x_train,y_train)
xg.score(x_test,y_test)
```

```python
pd.to_pickle(rf,'rf_99.pickle')
```

##### DISPLAYING RESULTS ######

```python
from sklearn.metrics import confusion_matrix as cm
y_pred = rf.predict(x_test)
print(cm(y_test,y_pred))
print('SCORE : ',rf.score(x_test,y_test))
sns.heatmap(cm(y_test,y_pred),annot=True,fmt='d')
```
